```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output
csvdata = pd.read_csv("SAIL2025_LVMA_data_3min_20August-25August2025_flow.csv")
data = (
    csvdata
    .assign(timestamp=lambda data: pd.to_datetime(data["timestamp"], format="%Y-%m-%d %H:%M:%S%z"))
    .sort_values(by="timestamp")
)
print(data)
Dataseries =  [col for col in data.columns[1:74]]


app = Dash(__name__)
```

```python
app.layout = html.Div(
    children=[
        html.H1(children="Pedestrian Analytics"),
        html.P(
            children=(
                "Pedestrians measured"
                " during SAIL 2025"
            ),
        ),
        html.Img(
            src="/assets/sensormap.png",  # path relative to the Dash server
            style={"width": "200px", "height": "auto"}  # optional styling
        ),
        html.Div(
            children=[
                html.Div(
                    children=[
                        html.Div(children="Dataseries", className="menu-title"),
                        dcc.Dropdown(
                            id="data-filter",
                            options=[
                                {"label": data, "value": data}
                                for data in Dataseries
                            ],
                            value="CMSA-GAKH-01_0",
                            clearable=False,
                            className="dropdown",
                        ),
                    ]
                ),
            ],
        ),
        html.Div(
            children=[
                html.Div(
                    children=dcc.Graph(
                        id="pedestrian-count",
                        config={"displayModeBar": False},
                    ),
                    className="card",
                ),
            ],
        ),
        html.Div(
            children=[
                html.Div(
                    children=dcc.Loading(dcc.Graph(id="pedestrian-count-animated"), type="cube")
                ),
            ],
        ),
    ],
)
    
```
